```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

```

# Hat 0.005 cifar

```python
df = pd.read_csv('/home/cp/self/ml/hat/res/good_exp/hat_500_0.0005_xavier_adam_cifar/train_loss_app_hat_exp_cifar_opt_adam_wt_init_xavier_lr_0.0005_nepoch_500.csv')

# Ask the user for the column to use for lines
# line_column = input("Enter the column name for lines: ")
line_columns = ['Train_loss', 
#                 'Train_acc',
#                 'Avg_train_reg',
       'Val_loss',
#                 'Val_acc',
#                 'Avg_val_reg', 
#                 'Learning_rate'
               ]
line_columns2 = [
#     'Train_loss', 
                'Train_acc',
#                 'Avg_train_reg',
#        'Val_loss',
                'Val_acc',
#                 'Avg_val_reg', 
#                 'Learning_rate'
               ]

# Create subplots for each task (category)
fig = make_subplots(rows=10, cols=2, subplot_titles=[f'Task {i}' for i in range(10)], shared_xaxes=True)

for category in range(10):  # Assuming categories range from 0 to 9
    # Filter the DataFrame for the current category
    category_df = df[df['Task'] == category]
    colors = ["Blue","Red"]
    for i, column in enumerate(line_columns):
        # Add a line trace to the subplot
        fig.add_trace(
            go.Scatter(x=category_df["Epochs"], y=category_df[column], mode='lines', name=column,line=dict(color=colors[i])),
#             row=(category // 2) + 1,  # Calculate the row position
#             col=(category % 2) + 1,   # Calculate the column position
            col=1,
            row=category+1,
            
        )
    colors = ["Black","Green"]
    
    for i, column in enumerate(line_columns2):
        # Add a line trace to the subplot
        fig.add_trace(
            go.Scatter(x=category_df["Epochs"], y=category_df[column], mode='lines', name=column, line=dict(color=colors[i])),
#             row=(category // 2) + 1,  # Calculate the row position
#             col=(category % 2) + 1,   # Calculate the column position
            col=2,
            row=category+1,
            
        )

# Update layout
fig.update_layout(height=2000, width=900, title_text="Line Plots for Each Task",
                 legend=dict(orientation="v", y=1, x=1),)

# Show the figure
fig.show()
```
